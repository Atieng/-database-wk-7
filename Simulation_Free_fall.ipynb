### The physical model

**simulate free-fall motion using the equation:** $y(t)=h_0-\frac{1}{2} g t^2$





```python
# Import the necessary Modules for that I will use for this experiment
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Declasre the variables for the line plot
h0=100.0  #metres
g=9.8     #Gravity in m/s^2
t =np.linspace(0,0.1,10)
y =h0-(0.5*g*(t**2))
print(y)
```

```python
# Plot the Line plot
plt.figure(figsize=(10,6))
plt.plot(t,y, linewidth =2, marker ="o",label="height y(t)");
plt.axvline(0,linewidth=1.5, label="Ground(y=0)", color ="red", linestyle="--")
plt.title("Free Fall Motion Graph")
plt.xlabel("Time (s)")
plt.ylabel("Height (m)")
plt.grid()
plt.legend();
```
